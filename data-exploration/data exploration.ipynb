## General code

```python
def open_data(line):
    
    df = pd.read_csv(line) 
    return df
```

```python
def describe_data(line):
    
    df = pd.read_csv(line) 
    return df.describe()
    
```

```python
def count_length(line):
    
    df = pd.read_csv(line) 
    return df.count()
```

```python
def are_there_nan_values(line):
    
    df = pd.read_csv(line) 
    df1 = df.copy()
    df1.dropna()
    
    if len(df) == len(df1):
        return "There are no NaN values in this dataset"
    if len(df) > len(df1):
        x = len(df) - len(df1)
        return f"There are {x} NaN values"
    else:
        return False 
```

```python
def dropna(line):
    
    df = pd.read_csv(line) 
    df.dropna()
    return df
```

```python
def plot_amount(line):
    
    df = pd.read_csv(line) 
    df.dropna()
    plot = sns.distplot(df['case AMOUNT_REQ'])
    return plot;
    
```

```python
def steps(line):
    
    df = pd.read_csv(line) 
    df.dropna()
    df['case REG_DATE'] = df['case REG_DATE'].astype('str')
    df_steps=df.groupby('case concept:name').count()
    plot = df_steps['eventID '].plot()
    return plot;
    
```

```python
def dataframe_steps(line):
    
    df = pd.read_csv(line) 
    df.dropna()
    df['case REG_DATE'] = df['case REG_DATE'].astype('str')
    df_steps=df.groupby('case concept:name').count()
    return df_steps
```

```python
steps('BPI_Challenge_2012-training.csv')
```
